```python
df.fillna('')

df = df.set_index(['vertrag', 'journal']).apply(lambda x: x.str.split(';').explode()).reset_index()  

df.issn.replace(' ', '', regex=True, inplace=True)
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df.issn = df.issn.replace('nan', '')
```

```python
duplicated_df = df[df[['vertrag', 'journal']].duplicated(keep=False)]
df = df.drop(index=duplicated_df[duplicated_df.issn == ''].index)
```

```python
df.shape[0]
```

```python
df.to_csv('oam_zeitschriftenlisten.csv', index=False, encoding='utf-8')
```
